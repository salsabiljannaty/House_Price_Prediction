```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Contoh load data (ubah ke path data kamu)
df = pd.read_csv("dataset/data_bersih.csv")  # ganti dengan nama file datasetmu

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Histogram Distribusi Harga Rumah')
plt.xlabel('Harga Rumah (Rp)')
plt.ylabel('Jumlah')
plt.grid(True)
plt.tight_layout()
plt.savefig('gambar/histogram_harga_rumah.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel')
plt.tight_layout()
plt.savefig('gambar/heatmap_korelasi.png', dpi=300)
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sqft_living', y='price')
plt.title('Scatter Plot Harga vs Luas Bangunan (sqft_living)')
plt.xlabel('Luas Bangunan (sqft)')
plt.ylabel('Harga Rumah (USD)')
plt.grid(True)
plt.tight_layout()
plt.savefig('gambar/scatter_harga_vs_luas.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Ganti 'bedroom_count' ke 'bedrooms' jika nama kolom aslinya adalah 'bedrooms'
sns.boxplot(x='bedrooms', y='price', data=df, palette='Set2')

plt.title("Boxplot Harga Rumah Berdasarkan Jumlah Kamar Tidur", fontsize=14)
plt.xlabel("Jumlah Kamar Tidur", fontsize=12)
plt.ylabel("Harga Rumah (Rp)", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("gambar/boxplot_harga_kamar.png", dpi=300)
plt.show()
```

```python
# Pilih fitur-fitur numerik utama
selected_features = [
    'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
    'floors', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated'
]

# Pastikan hanya kolom yang ada di DataFrame
features_in_df = [col for col in selected_features if col in df.columns]

# Buat pairplot
sns.pairplot(df[features_in_df])
plt.suptitle("Pairplot Fitur Numerik", y=1.02)
plt.savefig("pairplot_variabel.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'], color='skyblue')
plt.title('Boxplot Harga Rumah')
plt.xlabel('Harga (Rp)')
plt.grid(True)
plt.tight_layout()

# Simpan sebagai file gambar (misalnya PNG)
plt.savefig('gambar/boxplot_harga_rumah.png', dpi=300) 
```
